# 6

Find the limit:

lim(x, y) -> (0, pi/4)

(sec(x))(tan(y))

```python
f = smp.sec(x) * smp.tan(y)

# Compute the limit as x -> 0
lim_x = smp.limit(f.subs(y, smp.pi/4), x, 0)

# Compute the limit as y -> pi/4
lim_y = smp.limit(f.subs(x, 0), y, smp.pi/4)

print('Limit as x -> 0:')
display(lim_x)
print('Limit as y -> pi/4:')
display(lim_y)
```

For the functions w = 9x^2 - 5y^2, x = cos(t) and y = sin(t)

Then evaluate dw/dt at t = pi/4

```python
# Define the functions
x = smp.cos(t)
y = smp.sin(t)
w = 9*x**2 - 5*y**2

# Differentiate w with respect to t
dw_dt = smp.diff(w, t)

# Print the derivative
print("dw/dt: ")
display(dw_dt.trigsimp())

# Evaluate the derivative at t = pi/4
dw_dt_at_pi_4 = dw_dt.subs(t, smp.pi/4)
print("dw/dt at t = pi/4: ")
display(dw_dt_at_pi_4)
```

# 18

Draw a dependency diagram and write a chain rule formula for dz/ds where:

z = h(m, n)

m = k(s)

n = f(s)

```python
from sympy import symbols, Function, diff

# Define the symbols
s = symbols('s')

# Define the functions
k = Function('k')(s)
f = Function('f')(s)
h = Function('h')(k, f)

# Compute dz/ds
dz_ds = diff(h, s)

# Print the derivative
print("dz/ds: ")
display(dz_ds)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes
G.add_node('s')
G.add_node('m')
G.add_node('n')
G.add_node('z')

# Add edges
G.add_edge('s', 'm', label='dm/ds')
G.add_edge('s', 'n', label='dn/ds')
G.add_edge('m', 'z', label='dz/dm')
G.add_edge('n', 'z', label='dz/dn')

# Draw the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1500)

# Draw edge labels
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.show()
```

Draw a dependency diagram and write a chain rule formula for dz/dq and dz/dr given the functions below:

z = m(x, y)

x = h(q, r)

y = f(q, r)

```python
from sympy import symbols, Function, diff

# Define the symbols
q, r = symbols('q r')

# Define the functions
h = Function('h')(q, r)
f = Function('f')(q, r)
m = Function('m')(h, f)

# Compute dz/dq and dz/dr
dz_dq = diff(m, q)
dz_dr = diff(m, r)

# Print the derivatives
print("Partial derivative of z with respect to q: ")
display(dz_dq)
print("Partial derivative of z with respect to r: ")
display(dz_dr)
```
