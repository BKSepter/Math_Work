Give a geometric description of the set of points in space whose coordinates satisfy the given pairs of equations:

x^2 + y^2 + z^2 = 9; x = 0

Answer = The circle y^2 + z^2 = 9 in the yz-plane.

```python
# Sphere radius
r = 9

# Create an array of angles from 0 to 2*pi
theta = np.linspace(0, 2.*np.pi, 100)

# Convert polar coordinates to cartesian coordinates
# y = r*cos(theta), z = r*sin(theta)
y = r * np.cos(theta)
z = r * np.sin(theta)

# Plot the circle
plt.figure(figsize=(5,5))
plt.plot(y, z)
plt.xlabel('Y')
plt.ylabel('Z')
plt.title('Intersection of Sphere and Plane x=0')
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')  # to keep the circle round
plt.show()
```

# 4

Find the distance between points P_1 and P_2:

P_1(2, -3, -4), P_2(-3, 2, -11)

```python
# Numerical

# Define the coordinates of the two points
P1 = np.array([2, -3, -4])
P2 = np.array([-3, 2, -11])

# Calculate the difference between the two points
diff = P2 - P1

# Calculate the square of the difference
square_diff = diff**2

# Sum the squares and take the square root to get the distance
distance = np.sqrt(np.sum(square_diff))

distance
```

```python
# Symbolic

from sympy import Matrix, sqrt

# Define the coordinates of the two points
P1 = Matrix([2, -3, -4])
P2 = Matrix([-3, 2, -11])

# Calculate the difference between the two points
diff = P2 - P1

# Calculate the square of the difference
square_diff = diff.applyfunc(lambda x: x**2)

# Sum the squares and take the square root to get the distance
distance = sqrt(sum(square_diff))

distance
```

# 9

Find the angle between the vectors u = sqrt(5)i - 8j and v = sqrt(5)i + j - 3k

```python
import numpy as np

# Define the vectors
u = np.array([np.sqrt(5), -8, 0])
v = np.array([np.sqrt(5), 1, -3])

# Calculate the dot product of u and v
dot_product = np.dot(u, v)

# Calculate the magnitudes of u and v
magnitude_u = np.linalg.norm(u)
magnitude_v = np.linalg.norm(v)

# Calculate the angle between u and v
angle = np.arccos(dot_product / (magnitude_u * magnitude_v))

angle
```

Find the length and direction (when defined) of u * v and v * u

Answers:

The direction of u * v is (2/3)i + (1/3)j + (2/3)k

```python
# Define the vectors
u = np.array([9, -2, -8])
v = np.array([8, 0, -8])

# Calculate the cross products
uv = np.cross(u, v)
vu = np.cross(v, u)

# Calculate the magnitudes (lengths) of the cross products
uv_mag = np.linalg.norm(uv)
vu_mag = np.linalg.norm(vu)

# Normalize the cross products to find their directions
uv_dir = uv / uv_mag
vu_dir = vu / vu_mag

# Print the results
print(f"u x v = {uv}, with magnitude = {uv_mag}, direction = {uv_dir}")
print(f"v x u = {vu}, with magnitude = {vu_mag}, direction = {vu_dir}")
```

# 13

Find the area of the triangle determined by points P, Q, and R

Find a unit vector perpendicular to plane PQR

P(-2, 2, -1), Q(-1, 0, 2), R(0, -1, -2)

Answers:

The area of the triangle is (3 * sqrt(19)) / 2

The unit vectors perpendicular to plane PQR:

(+-) (1/(3 * sqrt(19))) (11i + 7j + k)

```python
# Numerical

# Define the points
P = np.array([-2, 2, -1])
Q = np.array([-1, 0, 2])
R = np.array([0, -1, -2])

# Calculate the vectors PQ and PR
PQ = Q - P
PR = R - P

# Calculate the cross product of PQ and PR to find a vector perpendicular to the plane
normal_vector = np.cross(PQ, PR)

# Calculate the magnitude (length) of the normal vector
normal_magnitude = np.linalg.norm(normal_vector)

# Normalize the normal vector to find its direction
normal_direction = normal_vector / normal_magnitude

# Calculate the area of the triangle (which is half the magnitude of the cross product)
triangle_area = normal_magnitude / 2

# Print the results
print(f"The area of the triangle is: {triangle_area}")
print(f"The unit vector perpendicular to the plane is: {normal_direction}")
```

```python
# Symbolic

# Define the points
P = Matrix([-2, 2, -1])
Q = Matrix([-1, 0, 2])
R = Matrix([0, -1, -2])

# Calculate the vectors PQ and PR
PQ = Q - P
PR = R - P

# Calculate the cross product of PQ and PR to find a vector perpendicular to the plane
normal_vector = PQ.cross(PR)

# Calculate the magnitude (length) of the normal vector
normal_magnitude = normal_vector.norm()

# Normalize the normal vector to find its direction
normal_direction = normal_vector / normal_magnitude

# Calculate the area of the triangle (which is half the magnitude of the cross product)
triangle_area = normal_magnitude / 2

# Print the results
print(f"The area of the triangle is:")
display(triangle_area)
print(f"The unit vector perpendicular to the plane is:")
display(normal_direction)
```

Find the area of the parallelogram whose vertices are given below:

A(0, 0, 0)

B(3, 4, 5)

C(6, 2, 5)

D(3, -2, 0)

Answer:


The area of parallelogram ABCD is:

sqrt(649)

```python
# Define the points
A = Matrix([0, 0, 0])
B = Matrix([3, 4, 5])
C = Matrix([6, 2, 5])
D = Matrix([3, -2, 0])

# Calculate the vectors AB and AD
AB = B - A
AD = D - A

# Calculate the cross product of AB and AD to find a vector orthogonal to the parallelogram
orthogonal_vector = AB.cross(AD)

# The magnitude (length) of this orthogonal vector is equal to the area of the parallelogram
area = orthogonal_vector.norm()

# Print the result
print(f"The area of the parallelogram is:")
display(area)
```
