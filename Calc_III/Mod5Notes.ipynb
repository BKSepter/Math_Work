∫(-1, 3) ∫(0, 1) 18x^2 y^5 dy dx

```python
from scipy.integrate import dblquad

# Define the function to integrate
def f(y, x):  # note the order of the arguments
    return 18 * x**2 * y**5

# Define the limits of x
x_bounds = (-1, 3)

# Define the limits of y as functions (in this case constants)
def y_min(x):
    return 0

def y_max(x):
    return 1

# Compute the double integral
result, error = dblquad(f, x_bounds[0], x_bounds[1], y_min, y_max)

print("The result of the integral is:", result)
```

# 13

The integral gives the area of the region in the xy-plane. 

Sketch the region, label each bounding curve with its equation, and give the coordinates of the points where the curves intersect. 

Then find the area of the region:

∫(0, 12) ∫((y^2)/6, 2y) dxdy

```python
# Points of Intersection

from scipy.optimize import fsolve

# Define the functions
def f1(y):
    return y**2/6

def f2(y):
    return 2*y

# Define the function that represents the difference between f2 and f1
def diff(y):
    return f2(y) - f1(y)

# Use fsolve to find the y-values where f1(y) = f2(y)
y_values = fsolve(diff, [0, 12])  # [0, 12] are initial guesses

# Substitute the y-values into either of the original functions to find the x-values
x_values = f2(y_values)

# Print the points of intersection
for x, y in zip(x_values, y_values):
    print(f"The curves intersect at ({x:.2f}, {y:.2f})")
```

```python
# Graph

# Define the functions
def f1(y):
    return y**2/6

def f2(y):
    return 2*y

# Define the range of y
y = np.linspace(0, 12, 400)

# Plot the functions
plt.figure(figsize=(8, 6))
plt.plot(f1(y), y, label='x = y^2/6')
plt.plot(f2(y), y, label='x = 2y')

# Fill the area between the curves
plt.fill_betweenx(y, f1(y), f2(y), where=(f2(y) >= f1(y)), color='gray', alpha=0.5)

# Mark the points of intersection
plt.plot(f1(0), 0, 'ro')
plt.plot(f1(12), 12, 'ro')

# Annotate the points of intersection
plt.text(f1(0), 0, '  (0, 0)', verticalalignment='bottom')
plt.text(f1(12), 12, '  (24, 12)', verticalalignment='bottom')

# Set the labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Area between curves')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Evaluate the integral

from scipy.integrate import quad

# Define the functions
def f1(y):
    return y**2/6

def f2(y):
    return 2*y

# Define the function that represents the difference between f2 and f1
def integrand(y):
    return f2(y) - f1(y)

# Perform the integration over the interval [0, 12]
area, error = quad(integrand, 0, 12)

print("The area of the region is approximately {:.2f} square units.".format(area))
```

# 15

Find the average height of the paraboloid z = x^2 + y^2 over the square 0 =< x =< 4, 0 =< y =< 4

```python
from scipy.integrate import dblquad

# Define the function to integrate
def f(x, y):
    return x**2 + y**2

# Define the limits of x and y
x_bounds = (0, 4)
y_bounds = (0, 4)

# Perform the double integration
result, error = dblquad(f, x_bounds[0], x_bounds[1], lambda x: y_bounds[0], lambda x: y_bounds[1])

# Divide by the area of the region to get the average height
average_height = result / 16

print(f"The average height of the paraboloid is approximately {average_height:.2f} units.")

# Answer:
# 32/3
```

Describe the given region in polar coordinates:

The shaded area is a rectangle with coordinates on the xy-plane at (0, 3) and (3sqrt(3), 0)

There is an unshaded part of the rectangle in the shape of a quarter of a circle with coordinates on the xy-plane at (1, 0) and (0, 1)

Answer:

The lower portion of the region is 0 =< theta =< pi/6, 1 =< r =< 3sqrt(3)sec(theta)

The upper portion of the region is pi/6 =< theta =< pi/2, 1 =< r =< 3csc(theta)

```python
from sympy import symbols, sqrt, atan2, simplify

# Define the symbols
x, y = symbols('x y')

# Define the corners of the rectangle in Cartesian coordinates
x1, y1 = 0, 4
x2, y2 = 4*sqrt(3), 0

# Define the corners of the quarter-circle in Cartesian coordinates
x3, y3 = 2, 0
x4, y4 = 0, 2

# Convert the corners of the rectangle to polar coordinates
r1, theta1 = sqrt(x1**2 + y1**2), simplify(atan2(y1, x1))
r2, theta2 = sqrt(x2**2 + y2**2), simplify(atan2(y2, x2))

# Convert the corners of the quarter-circle to polar coordinates
r3, theta3 = sqrt(x3**2 + y3**2), simplify(atan2(y3, x3))
r4, theta4 = sqrt(x4**2 + y4**2), simplify(atan2(y4, x4))

print(f"Rectangle corners in polar coordinates: ({r1}, {theta1}), ({r2}, {theta2})")
print(f"Quarter-circle corners in polar coordinates: ({r3}, {theta3}), ({r4}, {theta4})")
```
